```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# parameters
omega1 = 2.0
omega2 = 2.5
K = 1.0

# system of equations
def kuramoto_2(t, theta):
    theta1, theta2 = theta
    dtheta1 = omega1 + K * np.sin(theta2 - theta1)
    dtheta2 = omega2 + K * np.sin(theta1 - theta2)
    return [dtheta1, dtheta2]

# time span
t_span = (0, 20)
t_eval = np.linspace(0, 20, 1000)

# initial conditions
theta0 = [0, 1]

# solve ODE
sol = solve_ivp(kuramoto_2, t_span, theta0, t_eval=t_eval)

# plot
plt.plot(sol.t, sol.y[0], label="theta1")
plt.plot(sol.t, sol.y[1], label="theta2")
plt.legend()
plt.xlabel("time")
plt.ylabel("phase")
plt.show()
```
